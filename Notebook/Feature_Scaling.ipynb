<b>What is Normalization? </b>

Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.



<b>What is Standardization?</b>

Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.

```python
new_df = pd.DataFrame(df[['Age','Tenure']])
#new_df = pd.DataFrame(df,columns = ['Age', 'Tenure'])      Same as above
new_df.head(5)
```

<b>This the internal formula of MinMaxScaler
$$
x' = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
$$


```python
scaler = MinMaxScaler()
normalized_df = scaler.fit_transform(new_df)
print(normalized_df)
```

<b>Task-2 STANDARDIZATION

$$

--------
Z-score Standardization Formula
$$
x′ = \frac{x - \mu}{\sigma}
$$
Where:  \
μ is the mean of the data   \
σ is the standard deviation


-------
1. Calculate the mean (μ)
$$
μ = \frac{2 + 3 + 5 + 6 + 6}{5} = \frac{22}{5} = 4.4
$$
-------
2. Calculate the standard deviation (σ)

Use **population standard deviation** (ddof = 0):

$$
σ= \sqrt{ \frac{(2-4.4)^2 + (3-4.4)^2 + (5-4.4)^2 + (6-4.4)^2 + (6-4.4)^2}{5} }
$$

$$
σ = \sqrt{ \frac{13.2}{5} } = \sqrt{2.64} \approx 1.6248
$$
---------
3. apply This Formula to each value
$$
x′ = \frac{x - 4.4}{1.6248}
$$
